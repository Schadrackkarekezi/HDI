#Linear Regression Model:

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# Data preparation
df_final['log_gross_inc_percap'] = np.log(df_final['gross_inc_percap'])

df_model = df_final[['hdi', 'life_expectancy', 'expec_yr_school', 'mean_yr_school', 'log_gross_inc_percap']]
df_model = df_model.dropna()

X = df_model[['life_expectancy', 'expec_yr_school', 'mean_yr_school', 'log_gross_inc_percap']]
y = df_model['hdi']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Evaluate
y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f'R²: {r2:.4f}')
print(f'MAE: {mae:.4f}')
print(f'RMSE: {rmse:.4f}')

# Coefficients
coefficients = pd.DataFrame({'Feature': X.columns, 'Coefficient': model.coef_})
print(coefficients)

# Plot predictions vs actual
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([0, 1], [0, 1], '--', color='red')
plt.xlabel('Actual HDI')
plt.ylabel('Predicted HDI')
plt.title('Actual vs Predicted HDI')
plt.title('Actual vs Predicted HDI')

plt.show()

# Residuals plot
residuals = y_test - y_pred
plt.figure(figsize=(8, 6))
sns.histplot(residuals, kde=True, color='purple')
plt.axvline(0, color='red', linestyle='--')
plt.xlim(-0.02, 0.02)
plt.title('Residuals Distribution')
plt.xlabel('Residuals')
plt.show()
```

#Identify and analyze countries where their happiness is either higher or lower than what would be expected from their HDI, by calculating and comparing standardized scores.

#Merges HDI and Happiness data, calculates the gap between standardized HDI and Happiness Score, and visualizes outliers in a scatter plot.

```python


from adjustText import adjust_text

# Load the HDI dataset
df = pd.read_csv('/content/drive/MyDrive/Data/hdr_general.csv', encoding='latin1')
df.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')

df_final = df[['country', 'year', 'hdi', 'life_expectancy', 'expec_yr_school', 'mean_yr_school', 'gross_inc_percap']].copy()

def classify_income(gni):
    if gni < 1145:
        return 'Low-Income'
    elif 1146 <= gni <= 4515:
        return 'Lower-Middle Income'
    elif 4516 <= gni <= 14005:
        return 'Upper-Middle Income'
    else:
        return 'High-Income'

df_final['income_group'] = df_final['gross_inc_percap'].apply(classify_income)

df_final['log_gross_inc_percap'] = np.log(df_final['gross_inc_percap'])

df_hdi_2016 = df_final[df_final['year'] == 2017].copy()

# Load the World Happiness Report dataset
happiness_df = pd.read_csv('/content/drive/MyDrive/Data/2016.csv', encoding='latin1')

# Merge HDI and Happiness data
merged_df = pd.merge(df_hdi_2016, happiness_df, left_on='country', right_on='Country', how='inner')

# Standardize HDI and Happiness Score to compare deviations
merged_df['HDI_zscore'] = (merged_df['hdi'] - merged_df['hdi'].mean()) / merged_df['hdi'].std()
merged_df['Happiness_zscore'] = (merged_df['Happiness Score'] - merged_df['Happiness Score'].mean()) / merged_df['Happiness Score'].std()

# Compute the gap between standardized HDI and Happiness Score
merged_df['HDI_Happiness_Gap'] = merged_df['Happiness_zscore'] - merged_df['HDI_zscore']

# Identify the most overperforming and underperforming countries
top_happier_than_expected = merged_df.nlargest(5, 'HDI_Happiness_Gap')[['country', 'hdi', 'Happiness Score', 'HDI_Happiness_Gap']]
top_less_happy_than_expected = merged_df.nsmallest(5, 'HDI_Happiness_Gap')[['country', 'hdi', 'Happiness Score', 'HDI_Happiness_Gap']]

# Scatter Plot with Highlighted Outliers
plt.figure(figsize=(12, 7))
sns.scatterplot(x=merged_df['hdi'], y=merged_df['Happiness Score'], alpha=0.6, label="All Countries", edgecolor='black')

# Highlight outliers where HDI does not correlate well with happiness
outliers = pd.concat([top_happier_than_expected, top_less_happy_than_expected])
sns.scatterplot(x=outliers['hdi'], y=outliers['Happiness Score'], color='red', s=150, label="Country", edgecolor='black')

# Annotate outliers with better visibility
texts = []
for _, row in outliers.iterrows():
    texts.append(plt.text(row['hdi'], row['Happiness Score'], row['country'], fontsize=12, ha='left', fontweight='bold', color='black', bbox=dict(facecolor='white', alpha=0.7, edgecolor='black')))

# Adjust text to avoid overlap
adjust_text(texts, arrowprops=dict(arrowstyle='-', color='black', lw=0.5))

plt.xlabel("HDI (Human Development Index)", fontsize=12)
plt.ylabel("Happiness Score", fontsize=12)
plt.title("HDI vs. Happiness Score (Countries Highlighted)", fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

```

#Creating Scatter Plots for Different Factors:

```python
# Function to create a clean bar chart with all factors
def create_updated_bar_chart(merged_df):
    # Define the factors to include
    factors = ['Freedom', 'Trust (Government Corruption)', 'Family', 'Generosity',
               'Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Dystopia Residual']

    # Get the top happier and less happy countries
    happier_countries = merged_df.nlargest(10, 'HDI_Happiness_Gap')
    less_happy_countries = merged_df.nsmallest(10, 'HDI_Happiness_Gap')

    # Calculate mean factor values
    happier_means = [happier_countries[factor].mean() for factor in factors]
    less_happy_means = [less_happy_countries[factor].mean() for factor in factors]

    # Create the bar chart
    fig, ax = plt.subplots(figsize=(14, 8))

    x = np.arange(len(factors))
    width = 0.35

    rects1 = ax.bar(x - width/2, happier_means, width, label='Happier Than Expected')
    rects2 = ax.bar(x + width/2, less_happy_means, width, label='Less Happy Than Expected')

    # Add labels and title
    ax.set_ylabel('Mean Factor Value')
    ax.set_title('Happiness Factors Comparison')
    ax.set_xticks(x)
    ax.set_xticklabels([f.replace(' (', '\n(') for f in factors], rotation=0)
    ax.legend()

    # No horizontal grid lines as requested
    ax.grid(False)

    plt.tight_layout()
    plt.show()

    # Return the data for potential further analysis
    return {
        'factors': factors,
        'happier_means': happier_means,
        'less_happy_means': less_happy_means
    }

# Function to create all seven scatter plots with trend lines and no country labels
def create_all_scatter_plots(merged_df):
    # Define all factors to plot
    factors = ['Freedom', 'Trust (Government Corruption)', 'Family', 'Generosity',
               'Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Dystopia Residual']

    for factor in factors:
        plt.figure(figsize=(12, 7))

        # Define the top happier and less happy countries
        happier_countries = merged_df.nlargest(10, 'HDI_Happiness_Gap')
        less_happy_countries = merged_df.nsmallest(10, 'HDI_Happiness_Gap')

        # Plot all countries
        sns.scatterplot(x=merged_df['hdi'], y=merged_df[factor],
                       alpha=0.6, color='gray', label="All Countries")

        # Plot happier than expected countries
        sns.scatterplot(x=happier_countries['hdi'], y=happier_countries[factor],
                       color='green', s=100, label="Happier Than Expected")

        # Plot less happy than expected countries
        sns.scatterplot(x=less_happy_countries['hdi'], y=less_happy_countries[factor],
                       color='red', s=100, label="Less Happy Than Expected")

        # Add a best fit line for all countries to show the general trend
        sns.regplot(x=merged_df['hdi'], y=merged_df[factor],
                   scatter=False, color='blue', line_kws={'linestyle':'--'})

        # Add labels and title
        plt.xlabel("Human Development Index (HDI)", fontsize=12)
        plt.ylabel(f"{factor} Score", fontsize=12)
        plt.title(f"HDI vs {factor}", fontsize=14, fontweight='bold')
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.legend()

        plt.tight_layout()
        plt.show()

# Example usage:
factor_data = create_updated_bar_chart(merged_df)
create_all_scatter_plots(merged_df)


```
